#### Now we have to check the relationship between the 'offence_code','offence_code_group' and 'name'
#### First we have to find out the unique values in these columns 

```python
offenceGroup = data['OFFENSE_CODE_GROUP'].unique().tolist()
print(len(offenceGroup))

offenceCode = data['OFFENSE_CODE'].unique().tolist()
print(len(offenceCode))

name = data['NAME'].unique().tolist()
print(len(name))
```

#### So from above we found out that we find that the 
 1. No null values are present in these so all are valid 
 2. 'offence_code' and 'name' has same unique values so there are unique offence_name for every offence_code
 3. 'offence_code_group' has less unique values so there are multiple offence_code in a every offence_group

Now we have to map the (offencecode with offencecodegroup) and (offencecode with name) 

```python
offenceGroupAndCode={}
offenceCodeAndName ={}
n =len(data)

for i in range(n):
    code = data["OFFENSE_CODE"].iloc[i]
    group = data["OFFENSE_CODE_GROUP"].iloc[i]
    name= data["NAME"].iloc[i]

    if group not in offenceGroupAndCode:
        offenceGroupAndCode[group] = [code]
    else:
        if code not in offenceGroupAndCode[group]:
            offenceGroupAndCode[group].append(code)

    if code not in offenceCodeAndName:
        offenceCodeAndName[code] = name
    
```

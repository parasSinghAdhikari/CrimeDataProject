```python
# Now we map the relationship between the street and its [lat and long]
streetLatLong = {}
n = len(data)
for i in range(0,n):
    street = data["STREET"].iloc[i]
    lat = data["Lat"].iloc[i]
    long =  data["Long"].iloc[i]
    
    if(pd.isna(street)):
        continue
    if(pd.isna(lat) or pd.isna(long) or lat==-1 or long==-1):
        continue   
        
        
    if street not in streetLatLong:
        streetLatLong[street] =[(lat,long)]
    else:
        latlonglist = streetLatLong[street]
        check=True
        for latlong in latlonglist:
            if latlong[0] == lat and latlong[1]==long:
                check=False
                break
        if(check):
            streetLatLong[street].append((lat,long)) 
streetLatLong
```

```python
# First we have to filling the values of (street) using (Lat and long)
def findStreet(lat,long):
    for key,latLongList in streetLatLong.items():
        for latlong in latLongList:
            if lat==latlong[0] and long==latlong[1]:
                return key
    return None    

for i in range(0,n):
    value = data.loc[i,"STREET"]
    if pd.isna(value):
        lat = data.loc[i,'Lat']
        long = data.loc[i,'Long']
        if pd.isna(lat) or lat==-1 :
            continue
        data.loc[i,'STREET'] = findStreet(lat,long)
```

```python
# mapping between reportarea and streets
street_reportNo= {}
n=len(data)
for i in range(n):
    num = data["REPORTING_AREA"].iloc[i]
    street = data["STREET"].iloc[i]
    
    if pd.isna(num):
        continue
    if pd.isna(street):
        continue
    
    if num not in street_reportNo:
        street_reportNo[num] = [street]
    else:
        if street not in street_reportNo[num]:
            street_reportNo[num].append(street)  
street_reportNo
```

```python
#so we have to fill missing values of  reportarea acoording to street
def findReportArea(street):
    for key,streetList in street_reportNo.items():
        if street in streetList:
            return key
    return None    

for i in range(0,n):
    value = data.loc[i,"REPORTING_AREA"]
    if pd.isna(value):
        street = data.loc[i,'STREET']
        if pd.isna(street) :
            continue
        data.loc[i,'REPORTING_AREA'] = findReportArea(street)
```

```python
# We map the report area and its(lat and long)
reportareatolatlong = {}
for i in range(len(data)):
    lat = data.loc[i, "Lat"]
    long = data.loc[i, "Long"]
    area = data.loc[i, "REPORTING_AREA"]

    if pd.notna(lat) and pd.notna(area) and lat != -1:
         if area not in reportareatolatlong:
             reportareatolatlong[area] = [(lat,long)]
         else:
             latlonglist = reportareatolatlong[area]
             check=True
             for latlong in latlonglist:
                 if latlong[0] == lat and latlong[1]==long:
                     check=False
                     break
             if(check):
                 reportareatolatlong[area].append((lat,long)) 
reportareatolatlong        
```

```python
# find the missing value of reportarea using the (lat and long)
def findreportarea(lat,long):
    for key,latLongList in reportareatolatlong.items():
        for latlong in latLongList:
            if lat==latlong[0] and long==latlong[1]:
                return key
    return None    

for i in range(0,n):
    value = data.loc[i,"REPORTING_AREA"]
    if pd.isna(value):
        lat = data.loc[i,'Lat']
        long = data.loc[i,'Long']
        if pd.isna(lat) or lat==-1 :
            continue
        data.loc[i,'REPORTING_AREA'] = findreportarea(lat,long)
```

```python
# now we map the district and reporting area
districtAndReportArea = {}
for i in range(n):
    district = data.loc[i,"DISTRICT"]
    reportArea = data.loc[i,"REPORTING_AREA"]

    if(pd.isna(reportArea)):
        continue
    if(pd.isna(district)):
        continue   
    if district not in districtAndReportArea:
        districtAndReportArea[district] = [reportArea]
    else:
        if reportArea not in districtAndReportArea[district]:
            districtAndReportArea[district].append(reportArea)     
districtAndReportArea
```

```python
# Now we fill the District the using the report area
def findDistrict(reportarea):
    for key,reportlist in districtAndReportArea.items():
        if reportarea in reportlist:
            return key
    return None    

for i in range(0,n):
    value = data.loc[i,"DISTRICT"]
    if pd.isna(value):
        reportarea = data.loc[i,'REPORTING_AREA']
        if pd.isna(reportarea) :
            continue
        data.loc[i,'DISTRICT'] = findDistrict(reportarea)
```

```python
temp = data[data["DISTRICT"].notna() & data["REPORTING_AREA"].notna() & data["STREET"].notna()]
temp.reset_index(drop=True,inplace=True)
```

```python
# relationship between the (lat ,long) with the unique values of (district,reportarea and,street) 
DRSlatlong = {}
n = len(temp)
for i in range(n):
    district,reportarea,street,lat,long = temp.loc[i,["DISTRICT","REPORTING_AREA","STREET","Lat","Long"]]

    if pd.isna(lat) or lat==-1:
        continue
    
    if (district,reportarea,street) not in DRSlatlong:
        DRSlatlong[(district,reportarea,street)] = [(lat,long)]
    else:
        latlonglist = DRSlatlong[(district,reportarea,street)]
        check=True
        for latlong in latlonglist:
            if latlong[0] == lat and latlong[1]==long:
                check=False
                break
        if(check):
            DRSlatlong[(district,reportarea,street)].append((lat,long)) 
DRSlatlong
```

```python
# we have to fill missing  lat and long using street or district or reporting area
incident_to_latlong= {}
def findlatlong(district,reportarea,street):
    key = (district, reportarea, street)
    if key in DRSlatlong:
        return random.choice(DRSlatlong[key])
    return None
   

n = len(data)
for i in range(n):
    incident,district,reportarea,street,lat,long = data.loc[i, ["INCIDENT_NUMBER","DISTRICT", "REPORTING_AREA", "STREET", "Lat", "Long"]]

    if pd.isna(lat) or lat==-1:
        if pd.isna(district) or pd.isna(reportarea) or pd.isna(street):
            continue
        if incident in incident_to_latlong:
            data.loc[i,"Lat"] = incident_to_latlong[incident][0]
            data.loc[i,"Long"] = incident_to_latlong[incident][1] 
        else:
            latlong = findlatlong(district,reportarea,street)    
            if(latlong==None):
                continue
            incident_to_latlong[incident] = latlong 
            data.loc[i, "Lat"] = latlong[0]
            data.loc[i,"Long"] = latlong[1] 
    
```
